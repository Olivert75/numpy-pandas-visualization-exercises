## Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
from env import host, user_name, password
def get_db_url(username,password,host,db_name):
    return f'mysql+pymysql://{user_name}:{password}@{host}/{db_name}'
    
chipotle_url = get_db_url(user_name,password,host, 'chipotle')
orders_df = pd.read_sql('select * from orders', chipotle_url)
```

```python
#Changing item_price from string to float and ignore the warning
orders_df['item_price'] = orders_df['item_price'].str.replace('$','',regex = False).str.replace(',','',regex = False).astype('float')
```

```python
orders_df
```

```python
orders_df.groupby('item_name').item_price.sum().sort_values(ascending = False).nlargest(4, keep='all')
```

```python
ch_top_four = orders_df.groupby('item_name')['quantity', 'item_price'].sum().sort_values(by = 'quantity', ascending = False).head(4)
ch_top_four
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x ='quantity', y ='item_price', data =ch_top_four, hue = ch_top_four.index)
plt.legend(loc ='upper left')
```
