```python
import matplotlib.pyplot as plt
import math
```

```python
plt.figure(figsize = (14,10))

#Set x and y values
x = range(1,30)

#y = 0n + 1
y = [0 * n + 1 for n in x]

#y = log(n)
y2 = [math.log(n) for n in x]

#y = n
y3 = [n for n in x]

#y = n  log(n)
y4 = [n * math.log(n) for n in x]

#y = n^2
y5 = [n ** 2 for n in x]

#y = 2^n
y6 = [2 ** n for n in x]

#y = n!
y7 = [math.factorial(n) for n in x]

#y =n^n
y8 = [n ** n for n in x] 

plt.plot(x,y, ls = '-', marker = '.', label = '$O(1)$', color = "m", alpha= .2)
plt.plot(x,y2, ls = '-', marker = "*", label = "$O(log n)$", alpha= .3)
plt.plot(x,y3, ls = '-', marker = 'X', label = '$0(n)$', color = 'g', alpha= .4)
plt.plot(x,y4, ls = '-', marker = 'X', label = r'$O(n log n)$', color = 'orange', alpha= .3)
plt.plot(x,y5, ls = '-', marker = '*', label = r'$O(n^2)$', color = 'pink', alpha= .5)
plt.plot(x,y6, ls = '-', marker = '+', label = r'$O(2^n)$', color = 'purple', alpha= .7)
plt.plot(x,y7, ls = '-', marker = '+', label = r'$O(n!)$', color = 'r', alpha= 1)
plt.plot(x,y8, ls = '-', marker = 'X', label = r'$O(n^n)$', color = 'c', alpha= .5)

plt.grid(True, ls = '--', color = 'gray',)
plt.title('Big O Notation', fontsize=30, fontweight='bold',
         bbox={'facecolor': 'lightblue', 'alpha': 0.5, 'pad': 5})

plt.xlim(1,15)
plt.ylim(0,30)

plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```
