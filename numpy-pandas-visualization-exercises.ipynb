Use matplotlib to plot the following equation:

y =x^2 − x + 2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
```

```python
x = range(-49, 50)
y = [num**2 - num + 2 for num in x]
plt.plot(x, y)
plt.title('$y = x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.annotate('The origin', xy=(0, 0), xytext=(15, 5), arrowprops={'facecolor': 'blue'})
plt.show()
```

Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y =√x

y = x^3

y = 2^x

y = 1 /(x + 1)

You can use functions from the math module to help implement some of the equations above.

```python
import math
```

```python
#y =√x

x = range (0,30)
y = [math.sqrt(num) for num in x]
plt.plot(x, y)
plt.title('$y =√x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
#y = x^3

x = range (0,30)
y = [num**3 for num in x]
plt.title('$y = x^3$')
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
#y = 2^x

x = range (0,10)
y = [2**num for num in x]
plt.title('$y = 2^x$')
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
#y = 1 /(x + 1)

x = range (0,10)
y = [1 /(num + 1) for num in x]
plt.title('$y = 1 /(x + 1)$')
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

Combine the figures you created in the last step into one large figure with 4 subplots.

Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
#Create variables
x = range (0,20)
y = [math.sqrt(num) for num in x]
y2 = [num**3 for num in x]
y3 = [2**num for num in x]
y4 = [1 /(num + 1) for num in x]

#Subplot row_num = 1, column_num = 2, active_num = 1, 2, 3 or 4
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('$y =√x$')
plt.subplots_adjust(hspace=1, wspace=5)

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('$y = x^3$')
plt.subplots_adjust(hspace=1, wspace=5)

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('$y = 2^x$')
plt.subplots_adjust(hspace=1, wspace=5)

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('$y = 1 /(x + 1)$')
plt.subplots_adjust(hspace=1, wspace=5)

plt.show()
```
